## 1번 문제(google)

1부터 10,000까지 8이라는 숫자가 총 몇번 나오는가?

8이 포함되어 있는 숫자의 갯수를 카운팅 하는 것이 아니라 8이라는 숫자를 모두 카운팅 해야 한다.
(※ 예를들어 8808은 3, 8888은 4로 카운팅 해야 함)

```
str(list(range(10001))).count('8')
```

```
''.join([str(i) for i in range(10001)]).count('8')
```

```
# 8이 하나라도 있는 경우 1개로 카운팅
count = 0
for i in range(10001):
    if '8' in str(i):
        count += 1
count
```

## 2번 문제(daum)

1차원의 점들이 주어졌을 때, 그 중 가장 거리가 짧은 것의 쌍을 출력하는 함수를 작성하시오. (단 점들의 배열은 모두 정렬되어있다고 가정한다.)

예를들어 S = [1, 3, 4, 8, 13, 17, 20] 이 주어졌다면, 결과값은 (3, 4)가 될 것이다.

```
s = [1, 3, 4, 8, 13, 17, 20]
m = max(s)
index = 0
for i in range(len(s) - 1):
    if m > s[i+1] - s[i]:
        index = i
        m = s[i+1] - s[i]

s[index], s[index+1]
```

```
import re

def dart(s):
    pat = re.compile(r'([0-9]|10)([SDT])([\*#]?)')
    temp = []
    proc = {'S': lambda x: x, 'D': lambda x: x*x, 'T': lambda x: x**3}
    for n, t, b in pat.findall(s):
        print(f'n : {n}, t : {t}, b : {b}')
        print(f'temp : {temp}')
        p = proc[t](int(n))
        if b == '#':
            p *= -1
        elif b == '*':
            p *= 2
            if temp:
                temp[-1] *= 2
        print(f'p : {p}')
        temp.append(p)
        print('----------------')
    return sum(temp)

testcase = ['1S2D*3T', '1D2S#10S']

for i in testcase:
    print(dart(i))
```
